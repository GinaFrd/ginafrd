```python
mod_amazon[mod_amazon["uniq_id"] == ("}")] #localizamos datos con errores
```

```python
mod_amazon.drop([2003, 2005], inplace = True)
```

```python
mod_amazon[mod_amazon['number_available_in_stock'].str.contains("used", na = False)].head() ##buscamos los casos de artículos usados
```

```python
item = ["new", "used"] #separamos usados y nuevos y nas
mod_amazon["item_type"] = mod_amazon['number_available_in_stock'].str.extract(f'({"|".join(item)})') ##?¿estudiar la función  extract?
mod_amazon.head(2)
```

```python
mod_amazon["item_type"].unique() 
```

```python
mod_amazon["price_new"] = mod_amazon["price"].str.split(" ", expand = True).get(0).str.split("£", expand = True).get(1)  #remover el signo libra
mod_amazon.head() #£3.42
mod_amazon[mod_amazon.price_new.str.contains(',', na=False)] #comprobar si el valor está en el rango de miles
mod_amazon["price_new"] = mod_amazon["price_new"].str.replace(",", "") #eliminar la coma
mod_amazon.head()

```

```python
mod_amazon["number_available_in_stock"] = mod_amazon["number_available_in_stock"].str.extract("(\d+)").astype(np.float) 
# extract the digits and change the type to float
mod_amazon.dtypes
```

```python
mod_amazon["number_of_reviews"] = mod_amazon["number_of_reviews"].str.extract("(\d+)").astype(np.float) #extraemos los numeros y los cambiamos de objeto a float
```
